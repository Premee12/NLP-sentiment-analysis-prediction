# Preprocessing and Modelling

To build the corpus from the 'verified_reviews' we perform the following - <br>
1. Replace any non alphabet characters with a space
2. Covert to lower case and split into words
3. Iterate over the individual words and if it is not a stopword then add the stemmed form of the word to the corpus

Positive words can be seen in the above word cloud - good, enjoy, amazing, best, great etc

```python
# create a collection of texts from ['verified_reviews'] using stemmer
corpus = []
stemmer = PorterStemmer()
for i in range(0, data.shape[0]):
  review = re.sub('[^a-zA-Z]', ' ', data.iloc[i]['verified_reviews'])
  review = review.lower().split()
  review = [stemmer.stem(word) for word in review if not word in STOPWORDS]
  review = ' '.join(review)
  corpus.append(review)
```

Using Count Vectorizer to create bag of words

```python
# Initialize the CountVectorizer with a maximum of 2500 features
cv = CountVectorizer(max_features = 2500)

#Store data into independent and dependent variables in X and y

# Fit the vectorizer to the corpus and transform the corpus into a document-term matrix
X = cv.fit_transform(corpus).toarray()
# Store the feedback data into the variable y
y = data['feedback'].values
```

max_features=2500 specifies the maximum number of features (unique words) to be considered. It will keep only the top 2500 most frequent words in the corpus. This helps in reducing the dimensionality of the data and can improve the performance of machine learning models by focusing on the most important features.

```python
#Saving the Count Vectorizer
pickle.dump(cv, open('Models/countVectorizer.pkl', 'wb'))
```

Checking the shape of X and y

```python
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
```

Splitting data into train and test set with 30% data with testing.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 15)

print(f"X train: {X_train.shape}")
print(f"y train: {y_train.shape}")
print(f"X test: {X_test.shape}")
print(f"y test: {y_test.shape}")
```

```python
print(f"X train max value: {X_train.max()}")
print(f"X test max value: {X_test.max()}")
```

We'll scale X_train and X_test so that all values are between 0 and 1.

```python
scaler = MinMaxScaler()

X_train_scl = scaler.fit_transform(X_train)
X_test_scl = scaler.transform(X_test)
```

```python
#Saving the scaler model
pickle.dump(scaler, open('Models/scaler.pkl', 'wb'))
```

#### Random Forest

```python
#Fitting scaled X_train and y_train on Random Forest Classifier
model_rf = RandomForestClassifier()
model_rf.fit(X_train_scl, y_train)
```

```python
#Accuracy of the model on training and testing data
 
print("Training Accuracy :", model_rf.score(X_train_scl, y_train))
print("Testing Accuracy :", model_rf.score(X_test_scl, y_test))
```

Note: from the difference between the training acc and testing acc it shows that there is an overfitting issue.

```python
#Predicting on the test set
y_preds = model_rf.predict(X_test_scl)
```

```python
y_preds
```

```python
#Confusion Matrix
cm = confusion_matrix(y_test, y_preds)
```

```python
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model_rf.classes_)
cm_display.plot()
plt.show()
```

K fold cross-validation

```python
accuracies = cross_val_score(estimator = model_rf, X = X_train_scl, y = y_train, cv = 10)

print("Accuracy :", accuracies.mean())
print("Standard Variance :", accuracies.std())
```

Applying grid search to get the optimal parameters on random forest (Hyperparameter tunining)

```python
params = {
    'bootstrap': [True],
    'max_depth': [80, 100],
    'min_samples_split': [8, 12],
    'n_estimators': [100, 300]
}
```

```python
cv_object = StratifiedKFold(n_splits = 2)

grid_search = GridSearchCV(estimator = model_rf, param_grid = params, cv = cv_object, verbose = 0, return_train_score = True)
grid_search.fit(X_train_scl, y_train.ravel())
```

```python
#Getting the best parameters from the grid search


print("Best Parameter Combination : {}".format(grid_search.best_params_))
```

```python
print("Cross validation mean accuracy on train set : {}".format(grid_search.cv_results_['mean_train_score'].mean()*100))
print("Cross validation mean accuracy on test set : {}".format(grid_search.cv_results_['mean_test_score'].mean()*100))
print("Accuracy score for test set :", accuracy_score(y_test, y_preds))
```

```python
import shap


# Generate feature names based on the number of columns in X_test_scl
num_features = X_test_scl.shape[1]
feature_names = [f"feature_{i}" for i in range(num_features)]

# Convert X_test_scl (NumPy array) to DataFrame
X_test_scl_df = pd.DataFrame(X_test_scl, columns=feature_names)

# Load the saved XGBoost model
model_xgb = pickle.load(open('Models/model_xgb.pkl', 'rb'))

# Initialize the SHAP explainer
explainer = shap.TreeExplainer(model_xgb)

# Calculate SHAP values
shap_values = explainer.shap_values(X_test_scl_df)

# Summary plot of SHAP values
shap.summary_plot(shap_values, X_test_scl_df)

# SHAP dependence plot for a specific feature (replace 'feature_0' with any actual feature name if known)
shap.dependence_plot('feature_0', shap_values, X_test_scl_df)

# SHAP force plot for a specific instance (e.g., the first instance in the test set)
shap.initjs()
shap.force_plot(explainer.expected_value, shap_values[0, :], X_test_scl_df.iloc[0, :])
```
